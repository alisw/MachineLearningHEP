```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfgen4_8 = pickle.load(open("AnalysisResultsGen_pt_cand4_8.pkl", "rb"))
dfgen4_8["zL"] = dfgen4_8["pt_cand"] / dfgen4_8["pt_jet"]
dfgen4_8_jetmin5 = dfgen4_8[dfgen4_8["pt_jet"]>5] 
dfgen4_8_jetmin10 = dfgen4_8[dfgen4_8["pt_jet"]>10]
dfgen4_8_jetmin20 = dfgen4_8[dfgen4_8["pt_jet"]>20]
dfgen4_8_jetmin40 = dfgen4_8[dfgen4_8["pt_jet"]>40]

plt.hist(dfgen4_8["pt_jet"], bins = 30, range = [0,100])
```

```python
fig, axes = plt.subplots(1,1, figsize=(10,10))
#axes[0].set_yscale("log")
axes.set_title("z-distribution")
axes.set_xlabel("pt D/pt jet")
axes.set_ylabel("dN/dz")
axes.hist(dfgen4_8_jetmin40["zL"], bins = 30, range = [0,1], label = "ptjet>5", histtype = "step", fill = False, color='b')
axes.hist(dfgen4_8_jetmin20["zL"], bins = 30, range = [0,1], label = "ptjet>20", histtype = "step", fill = False, color='y')
axes.hist(dfgen4_8_jetmin10["zL"], bins = 30, range = [0,1], label = "ptjet>10", histtype = "step", fill = False, color='r')
axes.hist(dfgen4_8_jetmin5["zL"], bins = 30, range = [0,1], label = "ptjet>20", histtype = "step", fill = False, color='g')
axes.set_ylim([1,1e6])
axes.legend()
```
