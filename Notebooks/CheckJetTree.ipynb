```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#dfgen4_8 = pickle.load(open("../AnalysisResultsGen_pt_cand4_8_1Mevents.pkl", "rb"))
dfgen4_8 = pickle.load(open("/data/HeavyFlavour/D0jetpp_20072019/mc/datapklskdecmerged77_2018/AnalysisResultsGen_pt_cand4_8.pkl", "rb"))
dfgen4_8["zL"] = dfgen4_8["pt_cand"] / dfgen4_8["pt_jet"]
dfgen4_8_jetmin5 = dfgen4_8[dfgen4_8["pt_jet"]>5] 
dfgen4_8_jetmin10 = dfgen4_8[dfgen4_8["pt_jet"]>10]
dfgen4_8_jetmin20 = dfgen4_8[dfgen4_8["pt_jet"]>20]
dfgen4_8_jetmin40 = dfgen4_8[dfgen4_8["pt_jet"]>40]
dfgen4_8_jetmin8 = dfgen4_8[dfgen4_8["pt_jet"]>8]
dfgen4_8_jetmin4 = dfgen4_8[dfgen4_8["pt_jet"]>4]

binsPtJet = [0,5.0,6.0,8.0,10.0,14.0,20.0,30.0,50.0]
plt.hist(dfgen4_8["pt_jet"], bins = binsPtJet, range = [0,200])
plt.yscale('log')
plt.ylabel("N of jets")
plt.xlabel("pT of jet")
```

```python
plt.hist(dfgen4_8["eta_jet"], bins = 20, range = [-1,1])
plt.ylabel("N of jets")
plt.xlabel("eta of jet")
```

```python
plt.hist(dfgen4_8["phi_jet"], bins = 36, range = [0,2*3.1415])
plt.ylabel("N of jets")
plt.xlabel("phi of jet")
```

```python
binsPtD = [0,4.0,5.0,6.0,7.0,8.0,10.0,12.0,16.0,24.0,36.0,]
plt.hist(dfgen4_8["pt_cand"], bins = binsPtD, range = [0,10])
plt.yscale('log')
plt.ylabel("N of D cand.")
plt.xlabel("pT of D")
```

```python
binsZ = [0.0,0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6,0.65,0.7,0.75,0.8,0.85,0.9,0.95,1.00,1.05,1.1]
fig, axes = plt.subplots(1,1, figsize=(10,10))
#axes[0].set_yscale("log")
axes.set_title("z-distribution")
axes.set_xlabel("z = pt_D/pt_jet")
axes.set_ylabel("dN/dz")
axes.hist(dfgen4_8_jetmin40["zL"], bins = binsZ, range = [0,1], label = "pt_jet>40", histtype = "step", fill = False, color='b')
axes.hist(dfgen4_8_jetmin20["zL"], bins = binsZ, range = [0,1], label = "pt_jet>20", histtype = "step", fill = False, color='y')
axes.hist(dfgen4_8_jetmin10["zL"], bins = binsZ, range = [0,1], label = "pt_jet>10", histtype = "step", fill = False, color='r')
axes.hist(dfgen4_8_jetmin5["zL"], bins = binsZ, range = [0,1], label = "pt_jet>5", histtype = "step", fill = False, color='g')
axes.hist(dfgen4_8_jetmin8["zL"], bins = binsZ, range = [0,1], label = "pt_jet>8", histtype = "step", fill = False, color='k')
#axes.set_ylim([1,4e5])
axes.legend()
```
